```
from transformers import pipeline
```

```
classifier = pipeline("zero-shot-classification")
#classifier = pipeline("zero-shot-classification", device=0) # to utilize GPU
```

```
def get_informative_non_informative(sequence):
  #from transformers import pipeline
  
  candidate_labels = ["informative", "non informative"]
  hypothesis_template = "The sentence is {}."
  res = classifier(sequence, candidate_labels, hypothesis_template=hypothesis_template)
  print(res)
  return res
```

```
def getTopic(sequence):
  #from transformers import pipeline
  #classifier = pipeline("zero-shot-classification")
  #candidate_labels = ["suggestion", "bug", "improvement", "malfunction", "praise"]
  candidate_labels = ["improvement suggestion", "bug", "feature request", "Information enquiry", "praise", "Content Request", "fault or malfunction", "Feature information", "Criticism"]
  
  #candidate_labels = ["informative", "non informative"]
  hypothesis_template = "The sentence is {}."
  res = classifier(sequence, candidate_labels, hypothesis_template=hypothesis_template)
  print(res)
  return res
```

```
def getSentiment(sequence):
  # from transformers import pipeline
  # classifier = pipeline("zero-shot-classification")
  candidate_labels = ["strongly negative", "weakly negative", "neutral", "strongly positive", "weakly positive"]
  candidate_labels = ["negative", "neutral", "positive"]
  # res = classifier(sequence, candidate_labels)
  # print(res)
  #candidate_labels = ["positive", "negative"]
  hypothesis_template = "The sentiment of this review is {}."
  res = classifier(sequence, candidate_labels, hypothesis_template=hypothesis_template)
  print(res)
  return res
```
